```python
# 크기변환 출력영상을 기준으로 
import cv2
import numpy as np
import matplotlib.pyplot as plt

def scale_normal(img, scale_x = 1, scale_y = 1):
    height, width = img.shape
    height_new, width_new = int(height*scale_y), int(width*scale_x)
    
    _img = np.zeros((height_new, width_new), dtype = np.uint8) #500 -> 250
    
    for y in range(height_new):
        for x in range(width_new):
            try:
              _img[y, x] = img[int(y/scale_y), int(x/scale_x)]
            except:
              print('An exception occurred')
    return _img


img = cv2.imread('images/lena.jpg', cv2.IMREAD_GRAYSCALE)
result = scale_normal(img, 2, 2)

cv2.imshow('origin',img)
cv2.imshow('scale',result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#보간법 interpolation
def scale_nearest(img, scale_x = 1, scale_y = 1):
    height, width = img.shape
    height_new, width_new = int(height*scale_y), int(width*scale_x)
    
    _img = np.zeros((height_new, width_new), dtype = np.uint8) #500 -> 250
    
    for y in range(height_new):
        for x in range(width_new):
            try:
              _img[y, x] = img[round(y/scale_y), round(x/scale_x)]
            except:
              pass
              #print('An exception occurred')
    return _img


img = cv2.imread('images/lena_roi.jpg', cv2.IMREAD_GRAYSCALE)
result = scale_nearest(img, 3, 3)


cv2.imshow('origin',img)
cv2.imshow('scale',result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 보간법 -> 크기변환?  이미지 이동,  회전, 변환
# 선형 -> 가중치 

def scale_bilinear(img, scale_x=1, scale_y=1):
    height, width = img.shape
    img_ = np.zeros((int(height*scale_y), int(width*scale_x)), dtype=np.uint8)
    for y in range(int(height*scale_y)):
        for x in range(int(width*scale_x)):
            q = x/scale_x-int(x/scale_x)
            p = y/scale_y-int(y/scale_y)
            try:
                X = int(x/scale_x)
                Y = int(y/scale_y)
                value = (1-p)*(1-q)*img[Y, X] + p*(1-q)*img[Y+1, X] + (1-p)*q*img[Y, X+1] + p*q*img[Y+1, X+1]
                if value > 255:
                    img_[y, x] = 255
                else:
                    img_[y, x] = int(value)
            except:
                pass
    return img_

img = cv2.imread('images/lena_roi.jpg', cv2.IMREAD_GRAYSCALE)     
result=scale_bilinear(img,3,3)    
 

 
cv2.imshow('origin',img)
cv2.imshow('scale',result)         

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이미지 크기 줄이는 방법 : Max, Min

def maxpool2d(img):
    height, width = img.shape
    height_new, width_new = int(height/2), int(width/2)
    _img = np.zeros((height_new, width_new), dtype = np.uint8)
    for y in range(height_new):
        for x in range(width_new):
            try:
              _img[y, x] = np.max(img[2*y:2*y+2, 2*x:2*x+2])
            except:
              print('An exception occurred')
    return _img

def minpool2d(img):
    height, width = img.shape
    height_new, width_new = int(height/2), int(width/2)
    _img = np.zeros((height_new, width_new), dtype = np.uint8)
    for y in range(height_new):
        for x in range(width_new):
            try:
              _img[y, x] = np.min(img[2*y:2*y+2, 2*x:2*x+2])
            except:
              print('An exception occurred')
    return _img

def meanpool2d(img):
    height, width = img.shape
    height_new, width_new = int(height/2), int(width/2)
    _img = np.zeros((height_new, width_new), dtype = np.uint8)
    for y in range(height_new):
        for x in range(width_new):
            try:
              _img[y, x] = np.mean(img[2*y:2*y+2, 2*x:2*x+2])
            except:
              print('An exception occurred')
    return _img

img = cv2.imread('images/lena.jpg', cv2.IMREAD_GRAYSCALE)     
max = maxpool2d(img)
min = minpool2d(img)    
mean = meanpool2d(img) 
resize = cv2.resize(img,dsize=(250, 250), interpolation = cv2.INTER_AREA)

cv2.imshow('origin',img)
cv2.imshow('max', max)   
cv2.imshow('min', min)
cv2.imshow('mean', mean)  
cv2.imshow("Nearest", scale_nearest(img,0.5,0.5))
cv2.imshow('resize', resize) 

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#이동

def translate(img, distance = (0, 0)):
    height, width = img.shape
    _img = np.zeros(img.shape, dtype=np.uint8)

    for y in range(height):
        for x in range(width):
            X = x + distance[0]
            Y = y + distance[1]
            #경계선 검사 ->
            if(X < 0) | (X >= width) | (Y < 0) | (Y >= height): 
                continue
            _img[Y, X] = img[y, x]
    return _img
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
 
img = cv2.imread('images/lena.jpg', cv2.IMREAD_GRAYSCALE)
moved_img = translate(img, (100, 100)) 

cv2.imshow('origin', img)
cv2.imshow('moved_img',moved_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#회전

def rotate(img, deg = 30):
    height, width = img.shape
    _img = np.zeros(img.shape, dtype=np.uint8)
    center_x = int(width/2)
    center_y = int(height/2)
    theta = (np.pi*deg)/(180.0)

    for y in range(height):
        for x in range(width):
            X = int((x-center_x)*np.cos(theta) - (y-center_y)*np.sin(theta) + center_x)
            Y = int((x-center_x)*np.sin(theta) + (y-center_y)*np.cos(theta) + center_y)
            #경계선 검사 ->
            if(X < 0) | (X >= width) | (Y < 0) | (Y >= height): 
                continue
            _img[Y, X] = img[y, x]
    return _img
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
 
img = cv2.imread('images/lena.jpg', cv2.IMREAD_GRAYSCALE)
rotate = rotate(img, deg = 45) 

cv2.imshow('origin', img)
cv2.imshow('rotate', rotate)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#cv2x.getRotationMatrix2D()

img = cv2.imread('images/lena.jpg', cv2.IMREAD_GRAYSCALE)
rows, cols = img.shape
M = cv2.getRotationMatrix2D((rows/2, cols/2), angle=45, scale =1)

rotate_img = cv2.warpAffine(img, M, (rows,cols))

cv2.imshow('origin', img)
cv2.imshow('rotate', rotate_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
